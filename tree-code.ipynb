```python
# Importing all the required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# Chosen iris flower data as the target data
iris=datasets.load_iris()

# Transfer the data into a DataFrame table
table=pd.DataFrame(data=iris.data, columns=iris.feature_names)
table
iris.target_names
```

```python
# Training the model
x=table
y=iris.target
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
# Comparing Actual and predicted labels
dict={
    'Actual': y , 'Predicted': y_pred
}

pd.DataFrame(dict)
```

```python
# Checking the accuracy of Model
report= classification_report(y,y_pred)
print(f'Accuracy report is following:\n\n {report}')

# With accuracy of 1.00 means, model is 100% accurate
```

```python
# Plotting the confusion matrix chart
from sklearn.metrics import ConfusionMatrixDisplay
chart=confusion_matrix(y,y_pred)
cm_display=ConfusionMatrixDisplay(chart, display_labels=['setosa', 'vesicolor','virginica'])
cm_display.plot()
plt.show()
```

```python
# Plotting the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(15,15))
plot_tree(model, fontsize=10, filled=True, feature_names=iris.feature_names, class_names=list(iris.target_names), impurity=True, proportion=True)
plt.tight_layout()
plt.show()
```
